```python
class Algebra():

    # define
    def __init__(self, name):
        self.name = name
        self.operator = set()

    # unset
    def __del__(self):
        print(f'The algebra {self.name} has been unset.')

    def reset(self):
        print(f'The algebra {self.name} has been reset.')
        self.operator = set()

    def add_elements(self, elements):
        self.operator = self.operator.union(elements)
        return self.operator

    def delete_elements(self, elements):
        self.operator = self.operator.difference(elements)
        return self.operator
```

```python
alg1 = Algebra('alg1')
```

```python
alg1.add_elements({1, 2})
```

```python
alg1.delete_elements({1})
```

```python
alg1.reset()
vars(alg1)
```

```python
alg1.default(True)
```
